Introduction

Our project set out to determine what makes the best neighborhood in Pittsburgh. Instead of focusing on wealth or population, our group decided to define “best” in terms of health and environment. We believe that a healthy neighborhood reflects the overall well-being of the people who live there. To capture this, we combined three submetrics: air quality, access to parks, and access to healthy food. Together, these factors create an overall picture of which neighborhoods provide the healthiest living environment.

Each team member focused on one of these submetrics using real data from the Western Pennsylvania Regional Data Center (WPRDC). One member analyzed average air quality index values across monitoring sites, another looked at park availability per neighborhood, and the third measured access to healthy food options by counting the number of grocery and convenience stores. Using these datasets, we aimed to find which neighborhoods offer the cleanest air, most green space, and best access to food, which we felt represented the key components of a healthy lifestyle.

Our Metric

We defined our group’s “health metric” as a combination of three scores:

Air Quality Score based on the average air quality index value at each monitoring site (lower index values indicate cleaner air)

Park Count Score based on the total number of public parks per neighborhood

Healthy Food Access Score based on the number of grocery and convenience stores available in each area

Each submetric was normalized to ensure fair comparison and then combined into a final composite health score. The idea behind this metric is that neighborhoods with cleaner air, more parks, and more food options are likely to be healthier and more livable places overall.

Air Quality Analysis

The air quality dataset provided average index values for several monitoring sites across Pittsburgh. A lower index value represents cleaner air and a healthier environment. From our analysis, Lawrenceville 2, West Mifflin, and Flag Plaza (Central Business District) had the lowest average air quality index values, showing cleaner and healthier air. On the other hand, areas like Lawrenceville, Lincoln, and Liberty 2 had higher index values, suggesting more pollution and less ideal air quality.

Park Availability Analysis

To measure the availability of green space, we counted the number of parks within each Pittsburgh neighborhood. Access to parks is an important indicator of community health because it encourages outdoor activity, reduces stress, and improves overall quality of life. According to our data, East Liberty ranked highest with the most parks, followed by the Central Business District, Beechview, and Point Breeze. Neighborhoods like Mount Washington and the South Side Flats also performed well, showing that they offer residents plenty of opportunities for recreation and relaxation.

Healthy Food Access Analysis

We used the number of grocery and convenience stores in each neighborhood as a way to measure how easy it is for residents to access food and daily essentials. While convenience stores do not always provide the healthiest options, they still represent food accessibility. The Central Business District ranked highest in this category, followed by Bloomfield, Brookline, and the South Side Flats. These neighborhoods have the most opportunities for residents to easily access food, which is a key part of overall health.

Determining the Healthiest Neighborhood

After analyzing all three submetrics, we combined the results to find which neighborhood offered the best balance of clean air, green space, and food access. Based on our findings, the Central Business District scored the highest overall. It consistently performed well across categories, ranking in the top for air quality, park access, and food availability. East Liberty and Mount Washington also ranked highly, offering both outdoor space and strong access to resources.

Overall, our analysis suggests that the Central Business District is the healthiest neighborhood in Pittsburgh. It combines environmental cleanliness, recreational space, and access to essentials in a way that promotes a high quality of life for residents.
