```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

try:
    stores_df = pd.read_csv('data.csv')
    neighborhoods = gpd.read_file('Neighborhoods_.shp')

except FileNotFoundError:
    print("---!!! ERROR !!!---")
    print("File not found. Make sure 'data.csv' AND all 'Neighborhoods_' files are in your folder.")
    raise

stores_gdf = gpd.GeoDataFrame(
    stores_df, 
    geometry=gpd.points_from_xy(stores_df.Lon, stores_df.Lat),
    crs="EPSG:4326"
)

stores_gdf = stores_gdf.to_crs(neighborhoods.crs)

stores_with_neighborhoods = gpd.sjoin(stores_gdf, neighborhoods, how="inner", predicate='within')
type_col_name = 'Category'
neighborhood_col_name = 'hood'
store_type_name = 'Convenience Store' 

convenience_stores = stores_with_neighborhoods[
    stores_with_neighborhoods[type_col_name] == store_type_name
]

store_counts = convenience_stores.groupby(neighborhood_col_name).size()


final_rankings = store_counts.sort_values(ascending=False)


print(f"\n--- Top 10 Neighborhoods (HIGHEST '{store_type_name}' Count) ---")
print(final_rankings.head(10))
print("-------------------------------------------------------")

if final_rankings.empty:
    print(f"No stores found with the category '{store_type_name}'.")
else:
    top_15_neighborhoods = final_rankings.head(15)
    plt.figure(figsize=(10, 8))
    top_15_neighborhoods.plot(kind='barh', color='red') # Changed color to red
    plt.title(f'Top 15 Neighborhoods by HIGHEST Number of {store_type_name}s')
    plt.xlabel(f'Number of {store_type_name}s')
    plt.ylabel('Neighborhood')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

Central Buisness District or Bloomfield seem to be the best neighborhoods for indivudals with an active lifestyle because just the sheer amount of groccery/convenience stores within these neighborhoods. Theres a great variety of access to food so our individuals with healthy habits will thrive in these neighborhoods.
